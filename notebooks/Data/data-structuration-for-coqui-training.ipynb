```python
dataset_path = '../../data/wavs/'
```

```python
from pathlib import Path

dataset = list(Path(dataset_path).rglob(f'*.wav'))
dataset[0]
```

```python
print(f"There are {len(dataset)} recordings of male voice audios.")
```

```python
filenames = [str(path).split('/')[-1] for path in dataset]
filenames[0]
```

```python
import librosa, time
from tqdm import tqdm

def get_total_dataset_duration(dataset_path, filenames):

    duration = 0
    print('Duration calculation in progress...')
    for filename in tqdm(filenames):
        audio_path = dataset_path +filename
        duration  += librosa.get_duration(path=audio_path)
    
    total_duration = time.strftime('%H:%M:%S', time.gmtime(duration))
    print('Total duration of all audio files in the dataset:', total_duration)
```

```python
get_total_dataset_duration(dataset_path, filenames)
```

```python
import pandas as pd

metadata = pd.read_csv('tsv/voiceweb_clips.tsv', sep='\t')
metadata
```

```python
# Check if all filenames (ids) are in the metadata file
len(set(filenames) & set(metadata['original_sentence_id'])) == len(filenames)
```

```python
import re

def clean_str(text):
    
    text_in_parentheses_reg = r'\([\s\S]*\)'
    text_in_brackets_reg    = r'\[[\s\S]*\]'
    
    cleaned_txt = text.replace('\"', '')
    cleaned_txt = cleaned_txt.replace('|', '')
    cleaned_txt = re.sub(text_in_parentheses_reg, '', cleaned_txt)
    cleaned_txt = re.sub(text_in_brackets_reg, '', cleaned_txt)
    
    return cleaned_txt.strip()
```

```python
from tqdm import tqdm

with open('metadata.csv', 'w') as f:
    print('Metadata file generation in progress...')
    for path in tqdm(filenames):
        audio_label = metadata[metadata['original_sentence_id']==path]['sentence'].values[0]
        audio_label = clean_str(audio_label)
        
        line = path+'.wav|' + audio_label+'\n'
        f.write(line)
```

```python
generated_metadata = pd.read_csv('metadata.csv', sep='|', header=None)
generated_metadata.head()
```
