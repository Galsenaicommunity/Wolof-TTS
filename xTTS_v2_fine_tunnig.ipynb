## Formatage du dataset 
Comme dans le code de l'entraînement, nous choisissons le formateur `coqui` :
```

```
Nous devons donc organiser le dataset comme suit :
```
├── datasets/
│   ├── wavs/
│   │   ├── xxx.wav
│   │   ├── yyy.wav
│   │   ├── zzz.wav
│   │   └── ...
│   ├── metadata_train.csv
│   ├── metadata_eval.csv
```
Et les métadonnées comme ceci :
```
audio_file|text|speaker_name
wavs/xxx.wav|How do you do?|@X
wavs/yyy.wav|Nice to meet you.|@Y
wavs/zzz.wav|Good to see you.|@Z
```



````python
from datasets import load_dataset
ds = load_dataset("galsenai/wolof_tts")
````

````python
import os
import csv
import numpy as np
from scipy.io.wavfile import write
from tqdm import tqdm



output_dir = "dataset"
wavs_dir = os.path.join(output_dir, "wavs")
os.makedirs(wavs_dir, exist_ok=True)


metadata_train_path = os.path.join(output_dir, "metadata_train.csv")
metadata_eval_path = os.path.join(output_dir, "metadata_eval.csv")


def export_data(split, csv_path):

    with open(csv_path, mode='w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile, delimiter='|')
        writer.writerow(["audio_file", "text", "speaker_name"])  # En-tête

        for i, sample in enumerate(tqdm(ds[split], desc=f"Exporting {split} dataset")):
            audio_array = sample['audio']['array']
            sampling_rate = sample['audio']['sampling_rate']
            text = sample['text']
            gender = sample['gender']

            wav_filename = f"{i:06d}.wav"
            wav_path = os.path.join(wavs_dir, wav_filename)

            write(wav_path, sampling_rate, np.array(audio_array, dtype=np.float32))

            speaker_name = f"@{gender[0].upper()}"  # Exemple : @M ou @F
            writer.writerow([f"wavs/{wav_filename}", text, speaker_name])

export_data("train", metadata_train_path)
export_data("test", metadata_eval_path)

print("Export terminé.")
````

````python
import os
checkpoint_path = "checkpoints/GPT_XTTS_FT-October-24-2024_12+43AM-0000000/"
````

````python
import torch
import torchaudio
from tqdm import tqdm
from underthesea import sent_tokenize

from TTS.tts.configs.xtts_config import XttsConfig
from TTS.tts.models.xtts import Xtts


device = "cuda:0" if torch.cuda.is_available() else "cpu"


xtts_checkpoint = os.path.join(checkpoint_path, "best_model_18006.pth")
xtts_config =  os.path.join(checkpoint_path,"config.json")
xtts_vocab = "checkpoints/XTTS_v2.0_original_model_files/vocab.json"


# Load model
config = XttsConfig()
config.load_json(xtts_config)
XTTS_MODEL = Xtts.init_from_config(config)
XTTS_MODEL.load_checkpoint(config, checkpoint_path=xtts_checkpoint, vocab_path=xtts_vocab, use_deepspeed=False)
XTTS_MODEL.to(device)

print("Model loaded successfully!")
````

````python
from scipy.io import wavfile
import numpy as np


result = XTTS_MODEL.synthesize(
    "maa ngi lay jaajëfal bu baax",
    config,
    speaker_wav="/kaggle/input/wolof-dataset/dataset/wavs/audio_7.wav",
    gpt_cond_len=3,
    language="wo",
)


audio_signal = result['wav']
sample_rate = 24000
audio_signal = audio_signal / np.max(np.abs(audio_signal))

from IPython.display import Audio, display

display(Audio(audio_signal, rate=sample_rate))
````
